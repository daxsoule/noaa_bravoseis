## How to use this notebook

### Setup
1. Run the **setup cell** (imports + FIG_DIR) once at the start of every session.

### Reviewing a cluster
Each cluster has two sample types:
- **STRATIFIED** (30 events): Sampled evenly across distance-from-centroid quantiles (core → edge). Shows how signal quality degrades from cluster center to periphery.
- **RANDOM** (30 events): Pure random draw for an unbiased purity estimate.

Each sample is split into **3 batches of 10** to keep notebook memory manageable.

### Workflow per batch
1. **Run** the code cell to display 10 panels.
2. **Fill in the verdict table** below it:
   - **Verdict**: Accept or Reject
   - **Identified As**: What you think the signal is (T-phase, icequake, fin whale, vessel, unknown, etc.)
   - **Comments**: Anything notable — pick quality, contamination, interesting features
3. Move to the next batch. **Do not re-run old batches** — their images stay loaded.

### After finishing both samples for a cluster
1. Fill in the **cluster summary table** at the bottom of that cluster's section.
2. Tell Claude you're done — the next cluster's panels will be generated and added to the notebook.

### Memory management
- Only run **one batch at a time**. Previous batches retain their images until the kernel restarts.
- When moving to a new cluster, you can **restart the kernel** and re-run just the setup cell + current cluster cells if memory gets high.
