## Review

**Observations:**
- Peak freq: 2.0–4.4 Hz, mostly 2.4–3.9 Hz
- Short duration: 1.2–4.1 s (shorter than lowband_2's tonal signals)
- Low SNR: 3.0–5.8, mostly around 3–4
- Steep negative spectral slope: -1.2 to -2.3
- Impulsive waveforms — sharp onset, quick decay (NOT tonal like lowband_2)
- Panel 5 stands out: very high amplitude (SNR=22.9), broadband burst 1–12 Hz
- Most others: short impulsive pulses with energy concentrated at 2–4 Hz
- Spectrograms show transient broadband energy bursts, not persistent horizontal bands
- Dates: heavily clustered in Jan 12–13, 2019 (312/488 events)
- Moorings: M3 and M6 predominant

**Cruise report cross-reference:**
- Jan 12–13 events occur BEFORE seismic survey started (Jan 21)
- Jan 12: BRA28 deployed 12:30 UTC, BRA29 deployed 18:32–19:24 UTC
- Jan 13: BRA30, BRA31, BRA33 deployed
- Events start at 17:33 UTC Jan 12 — coincides with deployment operations
- M3 and M6 concentration matches moorings being deployed those days
- Most likely source: ship noise (thrusters, DP, winch) during deployment ops

**Verdict:** DISCARD. Deployment noise artifacts, not geologic signals. Even if some are real, safer to exclude than risk including ship noise in the catalogue.
