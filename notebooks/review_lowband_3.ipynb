## Review

**Observations:**
- Peak freq: 1.5–4.9 Hz, mostly 2.0–3.4 Hz
- Short duration: 1.3–3.3 s (median 2.2s)
- Low SNR: 3.0–4.3, threshold-level
- Steep negative spectral slope: -1.9 to -4.2
- Two sub-populations: tonal oscillations (like weaker lowband_2) and impulsive blips (like lowband_4)
- Panels 7, 8, 14, 15: higher amplitude, more broadband — possibly real seismic but minority
- M1 (1,371) and M3 (1,152) massively dominant (80%), virtually nothing on M4
- Peak Nov–Dec 2019 (austral late spring)
- Cruise check: 139 events during deployment, zero during shooting

**Verdict:** DISCARD. Grab-bag cluster of ambient noise features specific to M1 and M3 — mix of tonal noise and STA/LTA false triggers. Low SNR, no consistent morphology, strong mooring bias.
