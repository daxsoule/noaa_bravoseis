**Observations:**
- Very regular, quasi-sinusoidal waveforms at 2–5 Hz — highly tonal/monochromatic
- Steep negative spectral slope (-1.7 to -4.1) — energy sharply concentrated
- Long duration: 5–11 s (much longer than typical T-phases at this frequency)
- Spectrograms show persistent horizontal energy bands at 2–4 Hz with harmonics at ~5 Hz and ~7 Hz
- Dates: mostly Oct 2019 – Feb 2020 (austral spring/summer)
- Moorings: M1, M2, M3, M5 (distributed across array)
